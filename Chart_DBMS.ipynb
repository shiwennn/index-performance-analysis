```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Query Operation Performance Comparison
df = pd.DataFrame({
    'Attribute': ['channel_id', 'region_id', 'sales_date', 'total_amount'],
    'No Index (s)': [0.35, 0.42, 0.38, 0.45],
    'B-Tree (s)': [0.18, 0.22, 0.25, 0.20],
    'Bitmap (s)': [0.20, 0.15, 0.18, 0.22]
})

operations = ['Select', 'Insert', 'Update', 'Delete']  
b_tree = [0.2, 0.3, 0.25, 0.15]  
bitmap = [0.15, 0.35, 0.2, 0.1] 
x = np.arange(len(operations))
width = 0.35

plt.bar(x, b_tree, width, label='B-Tree')
plt.bar(x + width, bitmap, width, label='Bitmap')

plt.xlabel('Operation Type')
plt.ylabel('Execution Time (seconds)')
plt.title('Query Operation Performance Comparison')
plt.xticks(x, operations)
plt.legend()
plt.show()

# Average Query Execution Time by Index Type
plt.figure(figsize=(10,6))
bar_width = 0.25
index = range(len(df))

plt.bar([i - bar_width for i in index], df['No Index (s)'], width=bar_width, label='No Index', color='skyblue')
plt.bar(index, df['B-Tree (s)'], width=bar_width, label='B-Tree', color='orange')
plt.bar([i + bar_width for i in index], df['Bitmap (s)'], width=bar_width, label='Bitmap', color='green')

plt.xticks(index, df['Attribute'])
plt.xlabel('Attributes')
plt.ylabel('Average Query Execution Time (s)')
plt.title('Average Query Execution Time Comparison')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Index Storage Overhead Comparison
attributes = ['channel_id', 'region_id', 'sales_date', 'total_amount']
bitmap_size = [0.125, 1.875, 3.125, 3.000]
btree_size = [2.0, 2.0, 3.0, 2.0]

x = np.arange(len(attributes))
width = 0.35

plt.figure()
plt.bar(x - width/2, bitmap_size, width, label='Bitmap', color='steelblue')
plt.bar(x + width/2, btree_size, width, label='B-Tree', color='orange')

plt.xlabel('Attribute')
plt.ylabel('Index Size (MB)')
plt.title('Index Storage Overhead Comparison')
plt.xticks(x, attributes)
plt.legend()
plt.show()

# Index Creation Time Comparison
attributes = ['channel_id', 'region_id', 'sales_date', 'total_amount']
bitmap_time = [0.22, 0.07, 0.08, 0.28]
btree_time = [0.15, 0.14, 0.23, 0.16]

x = np.arange(len(attributes))
width = 0.35

plt.figure()
plt.bar(x - width/2, bitmap_time, width, label='Bitmap', color='seagreen')
plt.bar(x + width/2, btree_time, width, label='B-Tree', color='indianred')

plt.xlabel('Attribute')
plt.ylabel('Creation Time (seconds)')
plt.title('Index Creation Time Comparison')
plt.xticks(x, attributes)
plt.legend()
plt.show()
```
